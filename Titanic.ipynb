```python
data=data.drop(columns='Cabin',axis=1)
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
# now we can preprocess data for training model
# converting categorical value into numerical value
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Sex']=encoder.fit_transform(data['Sex'])
```

```python
encoder.transform(["male","female"])
```

```python
data.head(5)
```

```python
# Embarked column conversion
data['Embarked']=encoder.fit_transform(data['Embarked'])
```

```python
encoder.classes_
```

```python
data.tail(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(columns=(['Name','Ticket','PassengerId','Survived']),axis=1)
y=data['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

from sklearn.linear_model  import LogisticRegression
model_LR = LogisticRegression()
```

```python
model_LR.fit(x_train,y_train)
```

```python
predict_LR=model_LR.predict(x_test)
```

```python
from sklearn import metrics
accuracy_LR= metrics.accuracy_score(y_test,predict_LR)*100
```

```python
print("Accuracy of Logistic Regression Model :",round(accuracy_LR,2),"%")
```

```python
from sklearn.ensemble import RandomForestClassifier
moderl_RF=RandomForestClassifier()
```

```python
moderl_RF.fit(x_train,y_train)
```

```python
pred_rf=moderl_RF.predict(x_test)
```

```python
accuracy_rf=metrics.accuracy_score(y_test,pred_rf)*100
```

```python
print("Accuracy of Random Forest Model :",round(accuracy_rf,2),"%")
```
